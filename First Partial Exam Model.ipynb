Acá podemos importar una nueva librería para poder resumir con gráficos el comportamiento estadístico de una variable categórica como es el Estado Matricula

```python
import matplotlib.pyplot as plt
```

```python
temp = df["Estado Matricula"].value_counts()
df1 = pd.DataFrame({'labels': temp.index,
'values': temp.values})

labels = df1["labels"]
sizes = df1["values"]
parches, texts = plt.pie(sizes, shadow= True, startangle =90, pctdistance= 1.1, labeldistance = 1.2)
plt.legend(parches, labels, loc="best")
plt.axis("equal")
plt.tight_layout()
plt.show()
```
